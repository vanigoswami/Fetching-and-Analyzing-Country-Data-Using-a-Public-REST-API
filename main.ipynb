# **Task 1 — Fetch and Validate the Response**

```python
import requests

# 1. We use .strip() to ensure no accidental trailing spaces are present
url = "https://restcountries.com/v3.1/all".strip()

# 2. Add a User-Agent header to mimic a standard browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

try:
    # 3. Pass the headers into the request
    response = requests.get(url, headers=headers, timeout=10)

    response.raise_for_status()

    print(f"HTTP Status Code: {response.status_code}")
    countries_data = response.json()
    print(f"Success! Total number of countries retrieved: {len(countries_data)}")

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
    # If it's still a 400, the server might be having a bad day or rejecting the specific path
except Exception as err:
    print(f"An unexpected error occurred: {err}")
```

# **Task 2 — Extract and Display Key Fields**

```python
import pandas as pd

# Extracting specific fields from the JSON structure
extracted_data = []

for country in countries_data:
    data_point = {
        "common_name": country.get("name", {}).get("common"),
        "population": country.get("population"),
        "region": country.get("region"),
        "area": country.get("area")
    }
    extracted_data.append(data_point)

# Create DataFrame
df = pd.DataFrame(extracted_data)

# Display the first 10 rows
df.head(10)
```

# **Task 3 — Summarize the Data**

```python
# 1. Which region has the highest total population?
region_population = df.groupby("region")["population"].sum()
top_region = region_population.idxmax()
top_region_value = region_population.max()

# 2. Which country has the largest area?
largest_country_idx = df["area"].idxmax()
largest_country_name = df.loc[largest_country_idx, "common_name"]
largest_country_area = df.loc[largest_country_idx, "area"]

print("--- Summary Analysis ---")
print(f"Region with the highest total population: {top_region} ({top_region_value:,})")
print(f"Country with the largest area: {largest_country_name} ({largest_country_area:,} sq km)")
```
